```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
import matplotlib.pyplot as plt
from project_app.app_settings import orchard

population = orchard.create_initial_population()
print(f"Długość populacji: {len(population)}")

profit_lsts = []
for i in range(len(population)):
    print(i)
    _, _, _, profit_lst = orchard.simulated_annealing(T_start=3000, T_stop=1000, iterations_in_temp=100,
                                                                         epsilon=4, iterations_epsilon=10, alpha=0.99,
                                                                         neighbour_type=1, initial_sol=i, verbose=False)
    profit_lsts.append(profit_lst)
```

```python
print(len(profit_lsts))
for enum, lst in enumerate(profit_lsts):
    plt.figure(figsize=(8, 6), dpi=200)
    plt.plot(range(len(lst)), lst, linewidth=.6)
    plt.xlabel("Iteracja", fontsize=16)
    plt.ylabel("Zysk", fontsize=16)
    plt.title("Rozwiązanie początkowe numer: {}".format(enum))
    ymax = max(lst)
    xmax = lst.index(ymax)
    plt.plot(xmax, ymax, marker='o', markersize=3, label="x={}, y={:.2f}".format(xmax, ymax))
    plt.legend(loc='lower right')
    plt.savefig('visuals/SA/init_pop{}'.format(enum))
    plt.show()
```
